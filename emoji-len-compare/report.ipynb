Source data of emoji: https://gist.github.com/oliveratgithub/0bf11a9aff0d6da7b46f1490f86a71eb

```python
import pandas as pd

file_path = 'resources/result.csv'
df = pd.read_csv(file_path, delimiter=';')

df.head()
```

Analyzing the differences between Elasticsearch and Java

```python
diff_elastic_java = df[df['Elasticsearch'] != df['Java']]

diff_elastic_java[['Emoji', 'Elasticsearch', 'Java']]
print(len(diff_elastic_java))
```

Analyzing the overlap between Python and Elasticsearch

```python
same_python_elastic = df[df['Python'] == df['Elasticsearch']]

same_python_elastic[['Emoji', 'Python', 'Elasticsearch']]
```

Analyzing the differences between Elasticsearch and C#_with_surrogate

```python
diff_elastic_c_w_s = df[df['Elasticsearch'] != df['C#_with_surrogate']]

diff_elastic_c_w_s[['Emoji', 'Elasticsearch', 'C#_with_surrogate']]
print(len(diff_elastic_c_w_s))
```

Analyzing the differences between Elasticsearch and C#_without_surrogate

```python
diff_elastic_c_wo_s = df[df['Elasticsearch'] != df['C#_without_surrogate']]

diff_elastic_c_wo_s[['Emoji', 'Elasticsearch', 'C#_without_surrogate']]
```
