```python
import pandas as pd

registered_df = pd.read_csv('resources/registrated.csv', sep=';')
all_articles = registered_df['articles'].dropna().str.split(', ').explode().unique()
articles_set = set(all_articles)
```

```python
thr_all_df = pd.read_csv('resources/THR_ALL.csv', sep=';')

matching_rows = thr_all_df[thr_all_df['Product No.'].isin(articles_set)]
matching_row_numbers = matching_rows.index.tolist()
print("Номера строк с совпадающими 'Product No.' для THR_ALL:")
for row_num in matching_row_numbers:
    print(row_num)
```

```python
tkr_all_df = pd.read_csv('resources/TKR_ALL.csv', sep=';')

matching_rows = tkr_all_df[tkr_all_df['Product No.'].isin(articles_set)]
matching_row_numbers = matching_rows.index.tolist()
print("Номера строк с совпадающими 'Product No.' для TKR_ALL:")
for row_num in matching_row_numbers:
    print(row_num)
```

```python
ustar_df = pd.read_csv('resources/USTAR_II.csv', sep=';')

matching_rows = ustar_df[ustar_df['Product No.'].isin(articles_set)]
matching_row_numbers = matching_rows.index.tolist()
print("Номера строк с совпадающими 'Product No.' для USTAR_II:")
for row_num in matching_row_numbers:
    print(row_num)
```

```python
product_no_set = set(thr_all_df['Product No.'].dropna())
product_no_set.update(tkr_all_df['Product No.'].dropna())
product_no_set.update(ustar_df['Product No.'].dropna())

unmatched_articles = articles_set - product_no_set

print("Артикулы из 'registered.csv', отсутствующие в остальных файлах:")
for article in sorted(unmatched_articles):
    print(article)
```
